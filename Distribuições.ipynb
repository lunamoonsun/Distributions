# 1. Distribuição Normal

```python
import numpy as np
import seaborn as sns
import pandas as pd
from scipy import stats
```

## 1.1. Distribuição para variáveis contínuas 

As variáveis são numéricas e categóricas, as numéricas podem ser contínuas (números reais) e discretas (números inteiros). As categóricas são valores considerados "sim" ou "não", 0 ou 1, A ou B, e assim por diante.  

A distribuição normal também é considerada como distribuição gaussiana. 

### 1.1.1. Distribuição normal - Gaussian Distribution

```python
dados_normal = stats.norm.rvs(size=1000, random_state=1) # Criando uma variável com características de distribuição normal com números aleatórios
```

```python
dados_normal # Gerando 1000 números aleatórios 
```

```python
min(dados_normal), max(dados_normal)
```

```python
sns.distplot(dados_normal, hist=True, kde=True) # Mostra a frequência dos dados
```

#### Cálculos estatísticos utilizando as medidas de posição

```python
print(f"A Média da distribuição é: {dados_normal.mean()}") # Média dos valores 
print(f"A Mediana da distribuição é: {np.median(dados_normal)}") # Valor que está no meio
print(f"A Moda da distribuição é: {stats.mode(dados_normal)}") # Número que mais aparace
print(f"A Variância da distribuição é: {np.var(dados_normal)}") # Variância
print(f"O Desvio Padrão é: {np.std(dados_normal)}") # Desvio padrão em relação a média
```

```python
nums_direita = np.sum(((dados_normal >=0.9810041339322116 ) & (dados_normal <= 0.9810041339322116+1)))
print(f"Quantidade de elementos que estão a direita: {nums_direita}")
```

```python
nums_esquerda = np.sum(((dados_normal <=0.9810041339322116 ) & (dados_normal >= 0.9810041339322116-1)))
print(f"Quantidade de elementos que estão a esquerda: {nums_esquerda}")
```

## 1.2. Enviesamento

```python
from scipy.stats import skewnorm
```

```python
dados_normal = skewnorm.rvs(a=0, size=1000) # Gerando os dados aleatórios na estrutura de dist normal, o a=0 é sem enviesamento
```

```python
sns.distplot(dados_normal) # Verificando a distribuição normal
```

```python
print(f"A Média da distribuição é: {dados_normal.mean()}") # Média dos valores 
print(f"A Mediana da distribuição é: {np.median(dados_normal)}") # Valor que está no meio
print(f"A Moda da distribuição é: {stats.mode(dados_normal)}") # Número que mais aparace
print(f"A Variância da distribuição é: {np.var(dados_normal)}") # Variância
print(f"O Desvio Padrão é: {np.std(dados_normal)}") # Desvio padrão em relação a média
print(f"O Coeficiente de Variação é: {stats.variation(dados_normal)*100} %") # % de variacao de dados
```

## 1.3. Distribuição normal padronizada 

```python
dados_normal_padronizada = np.random.standard_normal(size=1000) # Gerando os dados da distribuição normal e padronizada
```

```python
min(dados_normal_padronizada), max(dados_normal_padronizada)
```

```python
sns.distplot(dados_normal_padronizada)
```

```python
print(f"Média após padronização tende a 0, vamos observar para os dados: {dados_normal_padronizada.mean()}")
print(f"O Desvio Padrão tende a 1, vamos observar para os dados: {np.std(dados_normal_padronizada)}")
```

```python
media = dados.mean()
media
```

```python
desvio_padrao = np.std(dados)
desvio_padrao
```

```python
# dados padronizados
dados_padronizados
```
